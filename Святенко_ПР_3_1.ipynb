4. Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum) Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий.

```python
data.columns = ['rank', 'youtuber', 'subscribers', 'video_views', 'category', 'title', 'uploads', 'country', 'abbreviation', 'channel_type', 'video_views_rank', 'country_rank', 'channel_type_rank', 'video_views_for_the_last_30_days', 'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings', 'highest_yearly_earnings', 'subscribers_for_last_30_days', 'created_year', 'created_month', 'created_date', 'gross_tertiary_education_enrollment_(%)', 'population', 'unemployment rate', 'urban_population', 'latitude', 'longitude']

data_miss = data.isna().sum()/len(data)*100
print(f"The amount of gaps in each column:\n", data_miss)

data.replace(to_replace = " ", value = np.NaN)
```

5. Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float 


```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
data[numeric_columns] = data[numeric_columns].astype(float)
print(data)
print(data.dtypes)
```

6. Ще раз перевірте кількість пропусків.

```python
data_miss = data.isna().sum()/len(data)*100
print(f"The amount of gaps in each column:\n", data_miss)
```

7. Перевірте, скільки унікальних країн представлено у колонці "Country"

```python
unique_countries = data['country'].nunique()
print(f"Count of unique countries:", unique_countries)
```

8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'): 

```python
data['video_views'].plot(kind='hist', title='Distribution of views')
```

9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)

```python
max_views = data['video_views'].max()
min_views = data['video_views'].min()
mean_views = data['video_views'].mean()

print(f'Max views = ', max_views)
print(f'Min views = ', min_views)
print(f'Mean views = ', mean_views)
```

10.Знайдіть країну, де найбільше відео було завантажено на YouTube 

```python
result = data.groupby('country')['uploads'].sum().reset_index()

most = result['uploads'].max()
country_with_most = result[result['uploads'] == most]['country'].values[0]

print(result)
print(f"\nCountry with the biggest amount of uploaded videos: {country_with_most}")
print(f"The amount of uploaded videos: {most}")
```

11. Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads.

```python
most_views = data.loc[data['uploads'].idxmax()]
name_of_channel_max = most_views['title']

least_views = data.loc[data['uploads'].idxmin()]
name_of_channel_min = least_views['title']

print(f"Name of channel with most uploaded videos: ", name_of_channel_max)
print(f"Name of channel with least uploaded videos: ", name_of_channel_min)
```
